## Darcy's law for Reservoir Pressure profile with python

```python
### We need to know various presssure at various radius in a steady state condition
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## re= External radius of reservoir  rwf= wellbore  radius  pwf = flowing bottom hole pressure h= thickness  q=flowrate  mu=viscosity



```python
re=3000
rwf=0.5
pwf=1500
h= 60
k= 150
q=200
mu=15
B=1
```

```python
## Creating radius of array between re and rwf
```

```python
r= np.linspace(rwf,re,500)
```

```python
r
```

```python
pressure=[]
for i in range(len(r)):
    p=pwf + (141.2*q*mu*B*(np.log(r[i]/rwf))/k/h)
    pressure.append(p)
```

```python
pressure
```

```python
plt.style.available
```

```python
plt.style.use('classic')
plt.figure(figsize=(10,6))
plt.plot(r,pressure)
plt.xlabel("r(ft)",size=20)
plt.ylabel("p(r),psi ",size=28)
plt.title("Reservoir Pressure Profile")
plt.grid(True)
plt.legend()
```
