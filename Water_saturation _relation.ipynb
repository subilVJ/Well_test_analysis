```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
### First create data for water saturation
```

```python
sw=np.linspace(0,1,100)
```

```python
sw
```

```python
## Now lets calculate realtive permeability of water (krw) using corelation principle  
```

```python
krw=sw**3
```

```python
krw
```

```python
## Lets calculate realative permeability of oil (kro)
```

```python
kro=(1-sw)**2
```

```python
kro
```

```python
## To put this data into one data frame
```

```python
df_relperm=pd.DataFrame({'Water _saturation':sw,'Rel Perm water':krw,'Rel perm oil':kro})
```

```python
df_relperm.columns.unique()
```

## Now if you notice when my water saturation is increasing my realtive permeability of water  is increasing and oil is decreasing


```python
df_relperm.plot()
```

```python
df_relperm.plot(subplots=True)
```

```python
## Step 1: ploting relative permeability of water with saturation(krw vs sw)
```

```python
plt.plot(sw,krw)
```

```python
## Step 2: kro vs sw
```

```python
plt.plot(sw,kro)
```

```python
# step 3=krw and kro vs sw
```

```python
plt.plot(sw,kro)
plt.plot(sw,krw)
plt.xlabel("Water saturation")
plt.ylabel("Relative Permeability")
```

```python
## Adding the gridlines
```

```python
plt.plot(sw,kro)
plt.plot(sw,krw)
plt.xlabel("Water saturation")
plt.ylabel("Relative Permeability")
plt.grid(True)
```

```python
## Adding legends to the plot
plt.plot(sw,kro,label="Water Rel Perm(krw)")
plt.plot(sw,krw,label="Oil Rel Perm(kro)")
plt.xlabel("Water saturation")
plt.ylabel("Relative Permeability")
plt.grid(True)
plt.legend(loc="lower right")
```

```python
## Change the size of the plot
plt.figure(figsize=(16,10))
plt.plot(sw,kro,label="Water Rel Perm(krw)")
plt.plot(sw,krw,label="Oil Rel Perm(kro)")
plt.xlabel("Water saturation")
plt.ylabel("Relative Permeability")
plt.grid(True)
plt.legend(loc="lower right")
```
